# Baseball Data Analysis

This dataset utilizes data  from 2014 Major League Baseball seasons in order to develop an algorithm that predicts the number of wins for a given team in 2015 season based on several different indicators of success. There are 16 different features that will be used as the inputs to the machine learning and the output will be a value that represents the number of wins

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/dsrscientist/Data-Science-ML-Capstone-Projects/master/baseball.csv")
df
```

    - Saves (SV)

Output Variable:

```python
target_variable = 'W'
feature_variable = ['R', 'AB', 'H', '2B', '3B', 'HR', 'BB', 'SO', 'SB', 'RA', 'ER',
       'ERA', 'CG', 'SHO', 'SV', 'E']

def num_plots(feature_name):
    fig,axs = plt.subplots(1,3,figsize=(15,7))
    a1 = sns.boxplot(x=df[feature_name],ax=axs[0])
    a2 = sns.distplot(df[feature_name],bins=20,kde=True,ax=axs[1])
    a3 = sns.scatterplot(data=df,x=feature_name,y=target_variable,ax=axs[2])
    plt.show()
    
for i in feature_variable:
    num_plots(i)
```

# Remove skewness by power transform

```python
from sklearn.preprocessing import power_transform
x_new = power_transform(x,method='yeo-johnson')
x_new
```

```python
pd.DataFrame(x_new).skew().sort_values(ascending=False)
```

```python
x = pd.DataFrame((x_new),columns = x.columns)
```

# Scaling :

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_scaled = sc.fit_transform(x)
```

# MultiColinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif['vif'] = [variance_inflation_factor(x_scaled,i) for i in range(x_scaled.shape[1])]
vif['Features'] = x.columns

vif
```

'ER' is negatively correlated and has high colinearity value so we can drop that column

```python
x_scaled=df.drop(['ER'],axis=1)
```

# Training the model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.metrics import accuracy_score

acc = 0
for i in range(0,1000):
    x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,random_state=i,test_size=.22)
    lr = LinearRegression()
    lr.fit(x_train,y_train)
    y_pred = lr.predict(x_test)
    temp = r2_score(y_test,y_pred)
    if temp > acc:
        acc = temp
        best_rstate = i
print("Accuracy : ",acc,"Random State : ",best_rstate)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,random_state=0,test_size=.22)
```

# Finding best model

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor
from sklearn.model_selection import cross_val_score,cross_val_predict
```

```python
lr = LinearRegression()
la =Lasso()
ri = Ridge()
en = ElasticNet()
kn = KNeighborsRegressor()
dtr = DecisionTreeRegressor()
rfr = RandomForestRegressor()
ada = AdaBoostRegressor()

model = [lr,la,ri,en,kn,dtr,rfr,ada]
```

```python
score = []
mean_squared_err = []
mean_absolute_err = []
r2 = []

for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    print("Accuracy_score : ", m," is",m.score(x_train,y_train))
    score.append(m.score(x_train,y_train))
    predm = m.predict(x_test)
    
    print("Mean Squared Error : ",mean_squared_error(y_test,predm))
    mean_squared_err.append(mean_squared_error(y_test,predm))
    print("Mean Absolute Error : ",mean_absolute_error(y_test,predm))
    mean_absolute_err.append(mean_absolute_error(y_test,predm))
    print("R2 SCORE: ",r2_score(y_test,predm))
    r2.append(r2_score(y_test,predm))
    print("\n")
```

Here LinearRegression and DecisionTreeRegressor is giving high accuracy of 100%. But we cant consider this as final model so we can further Cross Validate and hypertune to find best model.

# Cross Validation

```python
mean_score = []
std = []

for m in model:
    CV = cross_val_score(m,x,y,cv=5)
    print("Score of ",m," is : ",CV)
    print("Mean score is ",CV.mean())
    mean_score.append(CV.mean())
    print("Standard Deviation is ",CV.std())
    std.append(CV.std())
    print('\n')
```

we are having RandomForestRegressor with higher accuracy of 62%

# HyperTuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor

GBR = GradientBoostingRegressor()
parameters = {'learning_rate':[0.01,0.02,0.03,0.04],
             'subsample' : [0.9,0.5,0.2,0.1],
             'n_estimators':[100,150,200,250],
             'max_depth': [4,6,8,10]}

grid_GBR = GridSearchCV(estimator=GBR,param_grid=parameters,cv=2,n_jobs=1)
grid_GBR.fit(x_train,y_train)
```

```python
print("Results from Grid Search : ")
print("\n The best estimator across all search params :",grid_GBR.best_estimator_)
print("\n The best score for all searched params :",grid_GBR.best_score_)
print("\n The best parameters for all searched params : ",grid_GBR.best_params_)
```

```python
param_grid = [
    {"bootstrap": [True, False],
     "criterion": ["mse", "mae"],
     "n_estimators"      : [10,20,30,50,100],
     "max_features"      : ["auto", "sqrt", "log2"],
     "min_samples_split" : [2,4,8],
    }
]

rf = RandomForestRegressor(random_state=42)
```

```python
reg_grid = GridSearchCV(rf, param_grid, n_jobs=-1, cv=5)
```
